```python
import random
```

```python
class RandomData:
    def __init__(self,r_restaurant, r_order, k_restaurant, k_order, distance):
        self.restaurants = {}
        self.orders = []
        self.generate_data(r_restaurant, r_order, k_restaurant, k_order, distance)
    
    def generate_data(self, r_restaurant, r_order, k_restaurant, k_order, distance):
        restaurants, res_x, res_y = generate_restaurant(k_restaurant, r_restaurant)
        orders = generate_order(k_order, r_order, distance, res_x, res_y)
        self.restaurants = restaurants
        self.orders = orders
    
    def generate_restaurant(self,k,r):
        '''
        :type k: int
        :rtype: dict
        '''
        center_x = random.uniform(0, 10000)
        center_y = random.uniform(0, 10000)

        count = 0
        restaurants = {}
        dict_x = set([])
        dict_y = set([])
        while count < k:
            res_x = random.uniform(center_x - r, center_x + r)
            res_y = random.uniform(center_y - r, center_y + r)
            while (res_x in dict_x):
                res_x = random.uniform(center_x - r, center_x + r)
            while (res_y in dict_y):
                res_y = random.uniform(center_x - r, center_x + r)
            dict_x.add(res_x)
            dict_y.add(res_y)
            count += 1
            restaurants[count] = [(res_x,res_y)]

        return restaurants, center_x, center_y
    
    def generate_order(self,k_order, r_order, distance, res_x, res_y):
        center_x = res_x + distance
        center_y = res_y + distance

        count = 0
        orders = []
        while count < k_order:
            order_x = random.uniform(center_x - r_order, center_x + r_order)
            order_y = random.uniform(center_y - r_order, center_y + r_order)
            order = random.randint(1,k_order)
            orders.append([(order_x,order_y),order])
            count += 1

        return orders
    
    
```
